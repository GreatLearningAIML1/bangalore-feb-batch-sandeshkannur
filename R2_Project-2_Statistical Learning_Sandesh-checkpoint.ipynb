```python
Titan_data["OldScheme"]=Titan_data["Old Scheme (in thousands)"]*1000
Titan_data["NewScheme"]=Titan_data["New Scheme (in thousands)"]*1000
#NOTE : mean value calculated from the newly converted field
print( "Old Scheme Mean Value : {0}".format(Titan_data["OldScheme"].mean()))
print( "New Scheme Mean Value : {0}".format(Titan_data["NewScheme"].mean()))
```

```python
# Let  μ2 = Mean value of New Scheme.
# μ1 = Mean value of Old Scheme.
# μd = μ2 – μ1   

# H0: μd ≤ 5000  
# HA: μd > 5000


t_statistic, p_value = ttest_1samp(Titan_data["NewScheme"]-Titan_data["OldScheme"], 5000)
print("p-Value  = {0}".format(p_value))
```

Power of the test (5 points)

```python
# Calculating Power of Test
# Compute the difference in Means between 2 sample means and divide by pooled Standard Deviation 
# number of Observations/tuples
# Set the alpha value to 0.05 and alternative values 'two-sided' , 'larger' , 'smaller'
newScheme=Titan_data["NewScheme"]
oldScheme=Titan_data["OldScheme"]
effect_size=(np.mean(newScheme) - np.mean(oldScheme)) / np.sqrt(((30-1)*np.var(newScheme)+(30-1)*np.var(oldScheme)) / 30+30-2)
print(effect_size)
```

```python
# calculate the t test
alpha = 0.05
powerOfTest=ttest_power(effect_size, nobs=30, alpha=alpha, alternative='two-sided')
print("Power of Test Value: {0}".format(powerOfTest))
```
